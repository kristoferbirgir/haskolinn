```python
#Verkefni 12) Shell röðun:

def shell_sort(a):
    N = len(a)
    bil = N//2
    while bil > 0:
        for i in range(bil):
            for j in range(i+bil, N, bil):
                while j >= bil and a[j-bil] > a[j]:
                    a[j], a[j-bil] = a[j-bil], a[j]
                    j -= bil
        bil = bil//2
    return a

print(shell_sort([8,3,2]))
print(shell_sort([8,5,1,9,6,2,1,7,11,3]))
print(shell_sort([12,34,56,78,90,23,45,67,89,39,17,31]))
```

```python
#Verkefni 13) Skrá með íslenskum orðum:

from urllib.request import urlopen
def spegilord(string):
    return string == string[::-1]
f = urlopen("https://cs.hi.is/python/ord.txt")
listi_fjogur = []
line = f.readline().decode("utf-8").strip()
while line:
    if spegilord(line):
        listi_fjogur.append(line)
    line = f.readline().decode("utf-8").strip()
f.close()

count = 0
lina = ""
for ord in range(len(listi_fjogur)):
    if count == 10:
        print(lina + ",")
        lina = ""
        count = 0
    elif count == 0:
        lina = listi_fjogur[ord]
        count += 1
    elif 0 < count < 10:
        lina = lina + ", " + listi_fjogur[ord]
        count += 1
print(lina)
```

```python
#Verkefni 13) Skrá með íslenskum orðum:

from urllib.request import urlopen

def einn_serhljodi(ord):
    serhljodar = "a,á,e,é,i,í,o,ó,u,ú,y,ý,æ,ö"
    count = 0
    for char in ord:
        if char in serhljodar:
            count += 1
        return count == 1
    
f = urlopen("https://cs.hi.is/python/ord.txt")
listi_fimm = []
line = f.readline().decode("utf-8").strip()
while line:
    if einn_serhljodi(line):
        listi_fimm.append(line)
    line = f.readline().decode("utf-8").strip()
f.close()

count = 0
lina = ""
for ord in range(len(listi_fimm)):
    if count == 10:
        print(lina + ",")
        lina = ""
        count = 0
    elif count == 0:
        lina = listi_fimm[ord]
        count += 1
    elif 0 < count < 10:
        lina = lina + ", " + listi_fimm[ord]
        count += 1
print(lina)
```

```python
from urllib.request import urlopen

einkunnir = {}
with urlopen("https://cs.hi.is/python/einkunn.txt") as file:
    for line in file:
        nr, einkunn = line.decode("utf-8").strip().split()
        einkunnir[nr] = einkunn

nofn = {}
with urlopen("https://cs.hi.is/python/nofn.txt") as file_2:
    for line in file_2:
        nr, nafn = line.decode("utf-8").strip().split(maxsplit=1)
        nofn[nr] = nafn

sameinad = {}
for key in nofn.keys():
    if key in einkunnir:
        sameinad[key] = (nofn[key], einkunnir[key])

print("{:<10} {:<30} {:>10}".format("Númer", "Nafn", "Einkunn"))
print("----------------------------------------------------")
for nr, values in sameinad.items():
    print("{:<10} {:<30} {:>10}".format(nr, values[0], values[1]))

print("----------------------------------------------------")

haesta_einkunn = 0
besti_nemandi = []
for key in sameinad:
    if float(sameinad[key][1]) > haesta_einkunn:
        haesta_einkunn = float(sameinad[key][1])
        besti_nemandi = [sameinad[key][0], sameinad[key][1]]

print("Besti nemandinn er:", besti_nemandi[0], "með", besti_nemandi[1], "í einkunn!")
```
