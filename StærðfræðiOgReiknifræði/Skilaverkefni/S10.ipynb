```python
#Verkefni 38.C3) Kennitölur og vartöluprófun: 

def er_gild_kennitala(kennitala):
    vartala_sum = sum([int(str(kennitala)[i]) * [3, 2, 7, 6, 5, 4, 3, 2][i] for i in range(8)])
    afgangur = vartala_sum % 11
    if afgangur == 0:
        vartala = 0
    else:
        vartala = (11 - afgangur) % 10
        if vartala_sum - 11 == 1:
            return False
    if vartala == 1:
        return False
    return vartala == int(kennitala[8])

url = "https://cs.hi.is/python/kennitolur.txt"
response = requests.get(url)
text = response.content.decode('utf-8')
data = np.loadtxt(io.StringIO(text), delimiter=',', skiprows=1, dtype=str)

for i in range(data.shape[0]):
    kennitala = data[i, 2]
    if not er_gild_kennitala(kennitala):
        print(f"{data[i, 0]} ({kennitala}) ógild kennitala.")
```

```python
#Verkefni 38.C4) Kennitölur og vartöluprófun: 

def get_foedingardagur(kennitala):
    day = int(kennitala[0:2])
    month = int(kennitala[2:4])
    year_prefix = kennitala[4:6]
    century = None
    
    if year_prefix == "00":
        century = 1900
    elif year_prefix <= "36":
        century = 1900
    else:
        century = 2000
    year = century + int(year_prefix)
    
    months = ['janúar', 'febrúar', 'mars', 'apríl', 'maí', 'júní', 'júlí', 'ágúst', 'september', 'október', 'nóvember', 'desember']
    month_name = months[month - 1]

    return f"{day}. {month_name} {year}"

print(get_foedingardagur("2808983689"))

print(get_foedingardagur("2112018000"))

print(get_foedingardagur("2207863120"))
```
