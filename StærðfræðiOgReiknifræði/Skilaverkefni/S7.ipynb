```python
#Verkefni 25.2) Jafna bestu línu: 

import numpy as np
import matplotlib.pyplot as plt

(ár, hiti, úrkoma) = np.loadtxt("https://cs.hi.is/python/hiti-urkoma.txt").T

(a, b) = np.polyfit(hiti, úrkoma, deg=1)

xp = np.linspace(min(hiti), max(hiti))
yp = a*xp + b

plt.figure(figsize=(12,6))
plt.scatter(hiti, úrkoma,c='blue', label='X-ás er ársúrkoma og Y-ás er hiti')
plt.plot(xp, yp, c='green', label='Lína bestu jöfnu')
plt.xlabel('Hiti (°C)')
plt.ylabel('Ársúrkoma (mm)')
plt.title('Meðalárshiti og ársúrkoma')
plt.legend()
plt.show()
```

```python
#Verkefni 25.3) Jafna bestu línu: 

(X,Y,Z)= np.polyfit(ár, hiti, deg=2)
(a, b) = np.polyfit(ár, hiti, deg=1)

yp1 = X*xp**2+Y*xp+Z
xp = np.linspace(min(ár), max(ár))
yp = a*xp + b

plt.figure(figsize=(12,6))
plt.scatter(ár, hiti, c='blue', label='Ársúrkoma/ár')
plt.plot(xp, yp, c='green', label='Lína bestu jöfnu')
plt.plot(xp, yp1, c='orange', label='Besta parabóla')
plt.xlabel('ár')
plt.ylabel('Ársúrkomu (mm)')
plt.legend()
plt.title("Ársúrkoma miðað við ár.")
plt.show()
```

```python
#Verkefni 28) Kínverskt jöfnuhneppi: 

import numpy as np, numpy.linalg as la
A = np.array([[8, -1], [7, -1]])
b = np.array([3, -4])
X = la.solve(A,b)
print(f'Heildarkostnaður á hlutnum er {int(X[1])} og það voru {int(X[0])} manneskjur sem leggja inn í púkkið.')
```

```python
#Verkefni 29.1) Fjarlægð milli borga:

import math
l_reykjavik = math.radians(64)
b_reykjavik = math.radians(-22)
l_london = math.radians(51.5)
b_london = math.radians(0)
theta = abs(b_london - b_reykjavik)

print("Hornið á milli Reykjavíkur og London er:",math.degrees(theta),"°")
```

```python
#Verkefni 29.2) Fjarlægð milli borga:

r = 6370
L = theta * r
print("Lengd hringbogans sem spannar horn á milli Reykjavíkur og London er:", L, "km")
```
