Enn betri nlgun fst me samsettri trapisureglu sem notar 
 trapisur. Forriti hana og prfi me (*) og (**) og =4
. Kanni lka hve strt 
 arf a vera til a f (nstum) 7 rtta aukastafi, sbr. uppgefin rtt gildi. Hr eru formlurnar sem arf a nota samt skringarmynd:
 
Bilinu $[a,b]$ er skipt upp  $n$ hlutbil sem hvert hefur lengd $\Delta x = \dfrac{b-a}{n}$ Skiptipunktarnir eru: $a = x_0 < x_1 < x_2 < \cdots < x_n = b$ ar sem $x_i = a + i\Delta x$



 fst nlgunin: $\int_{a}^{b} f(x)\,dx
\approx T_n = \frac{\Delta x}{2}(f(x_0) + 2f(x_1) + 2f(x_2) +
\cdots + 2f(x_{n-1}) + f(x_n) )$



# Lausn:

```python
def trap(f,a,b,n):
    # heildi f fr a til b me samsettri trapisureglu og n trapisum
    dx = (b - a)/n
    sum = 0
    for i in range(1,n):
        xi = a + i*dx
        sum += 2*f(xi)
    (x0, xn) = (a, b)
    sum += f(x0) + f(xn)
    I = dx/2*sum
    return I

from math import sin, e
def g(x): return sin(x)/x
def h(x): return e**x

n = 4
Ig = trap(g, 1, 2, n)
Ih = trap(h, 0, 1, n)
print(f'Ig={Ig:.3f}, Ih={Ih:.3f}')

Rtt_g = 0.6593299
Rtt_h = 1.7182818
for n in range(1,5000):
    skekkja = abs(trap(h, 0, 1, n) - Rtt_h)
    if skekkja < 0.5e-7:
        print(n)
        break
```
